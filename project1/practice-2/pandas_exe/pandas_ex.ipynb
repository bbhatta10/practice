<!-- Ex.10... Most & least frequent MPAA Rating in the dataset in terms of occurrences
 -->


```python
# Most & least frequent MPAA Rating in the dataset in terms of occurrences


import pandas as pd
df = pd.read_csv('movies_2.csv')
df_clean = df.dropna(subset=['MPAA Rating'])
df_count = df_clean.value_counts('MPAA Rating')
df_count
most_freq = df_count.idxmax()
least_freq = df_count.idxmin()
print(f'MPAA Rating occurance {df_count}')
print(f'Most frequent MPAA rating is {most_freq}')
print(f'Least frequent MPAA rating is {least_freq}')

```

```python
# Most & least expensive Genre (take an average of all Budget measures grouped by Genre - use the groupBy() method)


import pandas as pd
df = pd.read_csv('movies_2.csv')
clean_df = df.dropna(subset=['Genre','Budget'])
# clean_df

calc_avg = clean_df.groupby('Genre')['Budget'].mean()
calc_exp = calc_avg.idxmax()
calc_les = calc_avg.idxmin()
# print('Most expensive movie genre is:', calc_avg.idxmax)
print('Most expensive movie genre is:', {calc_exp})
print('Less expensive genre is:',{calc_les})
```
